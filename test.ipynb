```python
# statistics for ML
df = pd.read_csv("salary_dataset.csv")
df = df.head()
print(display(df))

# get summary
df.describe()

# get mean
df[["Age", "Experience_Years", "Bonus_USD", "Work_Hours_Per_Week", "Performance_Score"]].mean() # too long
df.select_dtypes(include="number").drop(columns=["Employee_ID", "Joining_Year"]).mean() # smarter way

# get median
df.select_dtypes(include="number").drop(columns=["Employee_ID", "Joining_Year"]).median()

# get standard deviation
df.select_dtypes(include="number").drop(columns=["Employee_ID", "Joining_Year"]).std()

# get corr
df_corr = df.select_dtypes(include="number").drop(columns=["Employee_ID", "Joining_Year"]).corr()
df_corr["Performance_Score"]
```

Alright for the correlation part, i figured out that the higher the age -> the higher the salary but doesn't apply to bonus, therefore the work hours per week, performance score and experience years -> the higher the bonus.

```python
# Encoding categorical data (get_dummies, factorize)
df = pd.read_csv("salary_dataset.csv")
df = df.head()
print(display(df))

# one-Hot Encoding(get_dummies)
Gender_n = pd.get_dummies(df["Gender"], dtype=int)

# Label Encoding (Factorize)
codes, uniques = pd.factorize(df["Gender"])
print(codes)
print(uniques)

# add one-hot encoding to df
df = df.join(Gender_n)
df

# replace label encoding to df
df["Gender"] = codes
df
```

figured out one-hot encoding outputs 2 columns, so best to use df.join() to add it to the df

```python
# encoding challenge
df = pd.read_csv("customer_data.csv")
print(display(df))

# one-hot encode gender
pd.get_dummies(df["gender"], dtype=int)

# one-hot encode payment
pd.get_dummies(df["payment"], dtype=int)

# combine all encoded columns into df
gender_n  = pd.get_dummies(df["gender"])
df.join(gender_n)
df

# factorize payment
codes, uniques = pd.factorize(df["payment"])
codes
uniques

# replace/add codes for gender using factorize
codes, uniques = pd.factorize(df["gender"])
df["gender"] = codes # to replace
df
df["gender_n"] = codes # to add
df
```
