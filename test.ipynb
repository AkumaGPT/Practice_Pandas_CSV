```python
# Encoding categorical data (get_dummies, factorize)
df = pd.read_csv("salary_dataset.csv")
df = df.head()
print(display(df))

# one-Hot Encoding(get_dummies)
Gender_n = pd.get_dummies(df["Gender"], dtype=int)

# Label Encoding (Factorize)
codes, uniques = pd.factorize(df["Gender"])
print(codes)
print(uniques)

# add one-hot encoding to df
df = df.join(Gender_n)
df

# replace label encoding to df
df["Gender"] = codes
df
```

figured out one-hot encoding(get_dummies) outputs new columns, so best to use df.join() to add it to the df
and label encoding(factorize) gives one column(codes) which can be replaced/added to the df

```python
# encoding challenge
df = pd.read_csv("customer_data.csv")
print(display(df))

# one-hot encode gender
gender_dummies = pd.get_dummies(df["gender"], dtype=int)
df = df.join(gender_dummies)

# one-hot encode payment
payment_dummies = pd.get_dummies(df["payment"], dtype=int)

# combine all encoded columns into df
pd.concat([gender_dummies, payment_dummies], axis=1)

# factorize payment
codes, uniques = pd.factorize(df["payment"])
codes
uniques

# add codes for gender using factorize
codes, uniques = pd.factorize(df["gender"])
df["gender_n"] = codes # to add
df
```

```python
#  still on 100 days of python
df = pd.read_csv("salaries_by_college_major.csv")
df.head()
df.shape
df.columns
df.isna()
df.tail()
clean_df = df.dropna()
# print(display(clean_df.tail()))

# find major with highest starting salary
clean_df[clean_df["Starting Median Salary"] == df["Starting Median Salary"].max()]
# or get the index of row with the largest value and use loc to get the corresponding name
clean_df["Starting Median Salary"].idxmax()
clean_df["Undergraduate Major"].loc[43]
# or double square brackets
clean_df["Undergraduate Major"][43]
# or using .loc to retrieve an entire row
clean_df.loc[43]

# challenge

# get the college major with the highest mid career salary
clean_df["Mid-Career Median Salary"].idxmax() # returns the index of max value of each column
clean_df["Undergraduate Major"][8]
# get the salary eearned
clean_df["Mid-Career Median Salary"][8]

# get the college with lowest starting salary
clean_df["Undergraduate Major"].loc[clean_df['Starting Median Salary'].idxmin()]
# get how much they earn after university
clean_df["Starting Median Salary"][clean_df['Starting Median Salary'].idxmin()]

# get college major with lowest mid-career salary
clean_df["Undergraduate Major"][clean_df["Mid-Career Median Salary"].idxmin()]
# get how much people expect to earn with this degree
clean_df["Mid-Career Median Salary"][clean_df["Mid-Career Median Salary"].idxmin()]

# sorting values and adding columns
# low risk major - get the difference between the earnings of the 10th and 90th percentile
# clean_df["Mid-Career 90th Percentile Salary"] - clean_df["Mid-Career 10th Percentile Salary"]
# or
# clean_df["Mid-Career 90th Percentile Salary"].subtract(clean_df["Mid-Career 10th Percentile Salary"])
spread_col = clean_df["Mid-Career 90th Percentile Salary"].subtract(clean_df["Mid-Career 10th Percentile Salary"])
clean_df.insert(1, "spread", spread_col)
clean_df.head()

# sorting by lowest spread
low_risk = clean_df.sort_values(by="spread", ascending=False)
low_risk[["Undergraduate Major", "spread"]].head()

# get potential values in 90th percentile
potential_majors = clean_df.sort_values(by="Mid-Career 90th Percentile Salary", ascending=False)
potential_majors[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head()
```
